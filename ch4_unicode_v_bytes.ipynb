```python
s = 'café' # The str 'café' has four Unicode characters
len(s)  
```

```python
# Encode str to bytes using UTF-8 encoding
b = s.encode('utf8')
b # bytes literals have a b prefix
```

```python
len(b) # bytes b has five bygtes 8the code point for the accented e is encoded as two bytes in UTF-8
```

```python
# Decode bytes to str using UTF-8 encoding
b.decode('utf8')
```

##### Byte essentials

```python
cafe = bytes('café', encoding='utf_8') # bytes can be built from a str, given an encoding
cafe
```

```python
cafe[0] # Each item is an integer in range(256)
```

```python
cafe[:1] # Slices of bytes are also bytes - even slices of a single byte
```

```python
cafe_arr = bytearray(cafe)
cafe_arr # There is no literal syntax for bytearray: they are shown as bytearray() with a bytes literal as argument
```

```python
cafe_arr[-1:] # A slice of a bytearray is also a bytearray
```

```python
# Binary sequences have a class method that str doesn't have, called fromhex, which builds a binary sequence
# by parsing pairs of hex digits optionally separated by spaces
bytes.fromhex('31 4B CE A9')
```

```python
# Building a binary sequence from a buffer-like object is a low-level operation
# that may involve type casting
import array
numbers = array.array('h', [-2, -1, 0 ,1 , 2]) # Typecode 'h' creates an array of short integers (16 bits)
octets = bytes(numbers) # octets holds a copy of the bytes that make up numbers
octets # These are the 10 bytes that represent the 5 short integers
```
