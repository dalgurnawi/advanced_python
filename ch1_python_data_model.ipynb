```python
# Demonstrating power of implementing special methods, __getitem__ and __len__
import collections
Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2,11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank,suit) for suit in self.suits
                                        for rank in self.ranks]
    
    def __len__(self):
        return len(self._cards)
    
    def __getitem__(self, position):
        return self._cards[position]
    # __getitem__ delegates to the [] operator, meaning we can slice.
    # This also makes the deck iterable
```

Use of namedtuple to construct a simple class to represent individual cards. These objects are bundles of attributes with no custom methods, just like a database record.

```python
# Can use random.choice to get a random item from a sequence.
from random import choice
deck = FrenchDeck()
choice(deck)
```
